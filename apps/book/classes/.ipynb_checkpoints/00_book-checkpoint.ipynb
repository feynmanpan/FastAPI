```python
def bid_cycle(prefix='00', digits=2):
    max_n = int('9'*digits)
    for i in cycle(range(0, max_n+1)):
        bid = f'{prefix}{i:0{digits}}'
        yield bid
```

```python
bids = bid_cycle()
bids
```
